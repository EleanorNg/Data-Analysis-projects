## The dataset:
The datasets contains information about rental appartments in Berlin. 
One row is for one appartment. The dataset contains 6 columns.  
- Region: is the Berlin district, where the flat is located
- Condition: is the condition of the flat
- Rooms: The number of rooms the flat has
- Rent: Monthly rent for the flat
- Year_Construction: The year in which the house was build
- Space: How many square meters does the flat has?

## The Task:

1. How many apartments are there in Mitte in a 'well-kept' condition?
2. How many apartments are there in Kreuzberg built before 1900?
3. What are the average spaces of the apartments having 4 rooms?
4. Which is the most expensive region on average and what is the maximum rent for this region?

Bonus:

```
import pandas as pd
from scipy import stats
```

```
rental=pd.read_csv("https://raw.githubusercontent.com/juliandnl/redi_ss20/master/berlin_rental.csv")
```

```
rental.head()
```

# 1.How many apartments are there in Mitte in a 'well-kept' condition?

```
rental[(rental['Region']=="Mitte")&(rental['Condition']=="well_kept")]
```

```
rental[(rental['Region']=="Mitte")&(rental['Condition']=="well_kept")].shape[0]
```

There are 6 Apartments in Berlin-Mitte in in a 'well-kept'condition

# 2.How many apartments are there in Kreuzberg built before 1900?

```
rental.query('Region=="Kreuzberg"&Year_Construction<1900')
```

```
rental.query('Region=="Kreuzberg"&Year_Construction<1900').shape[0]
```

There are 3 apartments in Kreuzberg built before 1900

# 3.What are the average spaces of the apartments by the number of rooms?

```
x=rental['Space'].mean()
x
```

```
y=rental['Rooms'].sum()
y
```

```
x*y
```

```
rental.groupby('Rooms')[['Space']].mean()
```

The average spaces of the apartments by number of rooms is 164910.41324607324 

# 4.Which is the most expensive region on average and what is the maximum rent for this region?

```
rental.head()
```

```
rental.groupby('Region')[['Rent']].mean().sort_values(by="Rent",ascending=False)
```

```
rental.groupby('Region')[['Rent']].mean().sort_values(by="Rent",ascending=False).iloc[0]
```

```
rental.groupby('Region')[['Rent']].agg(['max'])
```

The most expensive Region on average in Berlin is Mitte and the maximum Rent is 14207.00 Euros 

# Bonus Question:
5. What are the min, max and average number of rooms of the apartments in Köpenick?


```
rental.groupby('Region')[['Rooms']].agg(['min','max','mean'])
```

```
rental[rental['Region']=='Köpenick']['Rooms'].agg(['min','max','mean']).round(2)
```

The min number of rooms of the apartments in Köpenick is 1,
The max number of rooms of the apartments in Köpenick is 4 and 
The average number of rooms apartments in Köpenick is 2.28.

```
#Transformations:	
#Is there a difference in average room space (Space divided by Rooms) for apartments built after 2000 versus apartments built before 2000?
rental["Average_Room_space"] = (rental.Space / rental.Rooms).round(1)
rental.head()
```

```
rental_After_2000 = rental[rental.Year_Construction > 2000]
rental_After_2000.head()
```

```
rental_Before_2000 = rental[rental.Year_Construction < 2000]
rental_Before_2000.head()
```

```
Sample_A = rental_After_2000.Average_Room_space
Sample_A.head()
```

```
Sample_B = rental_Before_2000.Average_Room_space
Sample_B.head()
```

```
print("Null_Hypothesis = (Sample_A = Sample_B)")
```

```
t_results = stats.ttest_ind(Sample_A, Sample_B)
t_results
```

```
alpha = 0.05
print(alpha)
```

```
if (t_results[1] < alpha):
  print("Sample_A and Sample_B are different")
```

```
#How many apartments were posted each month?
Yearly_Apartment_Posted = rental.groupby('Year_Construction')[["Rooms"]].count()
Yearly_Apartment_Posted.tail()
```

```
Monthly_Apartment_Posted = (Yearly_Apartment_Posted/12).round(1)
Monthly_Apartment_Posted.tail()
```
