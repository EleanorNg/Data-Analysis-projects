```python
# Task2 
# How expensive are the car makes in general
cars.groupby('Make')[['Price']].mean()
```

```python
cars.groupby('Make')[['Price']].mean().round(2)
```

```python
cars.groupby('Make')[['Price']].mean().round(2).sort_values('Price',ascending='False')
```

```python
#Most expensive car
cars.groupby('Make')[['Price']].mean().round(2).sort_values('Price',ascending='False').iloc[0]
```

```python
# Most expensive car brand
cars.groupby('Kind')[['Price']].agg(['mean','count']).round(1)
```

```python
# Most expensive car brand
cars.groupby('Kind')[['Price']].agg(['mean','max','min']).round(1)
```

```python
#Functions :
today= date.today()
cars['Age']=cars.Year.apply(lambda x:today.year-x)
```

```python
cars.head(5)
```

```python
def age_of_the_car(year):
  return today.year-year
  
  cars['Age']=cars.Year.apply(age_of_the_car) #if we do something with the rows then we use axis
```

```python
def col2(row):
  return row.Kms/row.Age

  cars['KM_under_1_year']=cars.apply(col2,axis=1)
```
